```python
#separate the feature and target
X = df.drop(columns='target',axis=1)
Y = df['target']
```

```python
print(X)
```

```python
print(Y)
```

spliting the data into training and testing data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Build the neaural network

```python
# importing tensorflow and Keras
import tensorflow as tf 
tf.random.set_seed(3)  #to get reprodusable results
from tensorflow import keras
```

```python
# setting up the layers of Neural Network

model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),# flatten is used to change the data being feed that is is a matrix(569,30 in our case) form into a single dimension array(for input shape we put in the number of columns in you datset)
                          keras.layers.Dense(20, activation='relu'),#(hidden layer)relu (rectified linear units)
                          keras.layers.Dense(2, activation='sigmoid')#(output layer)(specific it is equal to the number of classes in the target column)
])
```

Compiling a neural network in deep learning is a crucial step that prepares the model for training. Here are the main reasons why we compile a neural network:

1. Defining the Loss Function
The loss function measures how well the model's predictions match the actual targets. Compiling allows you to specify which loss function to use, which is essential for guiding the training process. Different tasks (e.g., regression vs. classification) require different loss functions.

2. Choosing an Optimizer
The optimizer is responsible for updating the model's weights based on the gradients computed during backpropagation. By compiling the model, you can select an appropriate optimizer (like Adam, SGD, or RMSprop) that will influence how effectively the model learns.

3. Setting Evaluation Metrics
Compiling also allows you to specify metrics that will be used to evaluate the model's performance during training and validation. Metrics such as accuracy, precision, and recall help you understand how well the model is performing.

4. Configuring Learning Parameters
You can set parameters related to the training process, like learning rate and decay, which are essential for controlling how the model learns.

5. Building the Model Graph
Compiling the model creates a computational graph that outlines how data flows through the network, which is essential for efficiently executing forward and backward passes during training.

6. Performance Optimization
Some libraries optimize the computation graph during the compilation step, which can lead to faster training and inference times.

```python
# compiling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',#(used for 0,1,2)(categorical_crossentropy is used for one note encoding eg [1,0,0] if dog,[0,1,0]  for cat etc)
              metrics=['accuracy'])
```

```python
# training the Neural Network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

Visualize accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

accuracy of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
